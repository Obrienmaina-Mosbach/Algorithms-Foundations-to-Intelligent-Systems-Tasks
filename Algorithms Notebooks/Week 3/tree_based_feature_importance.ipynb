```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier

dataCreate = datasets.load_breast_cancer()

X = pd.DataFrame(dataCreate.data, columns=dataCreate.feature_names)
y = dataCreate.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

```

```python
clf = DecisionTreeClassifier(criterion='gini')

# Fit the decision tree classifier
clf = clf.fit(X_train, y_train)
```

```python
# Print the feature importances
feature_importances = clf.feature_importances_
```

```python
feature_importances
```

```python
# Sort the feature importances from greatest to least using the sorted indices
sorted_indices = feature_importances.argsort()[::-1] #ascending order
sorted_feature_names = dataCreate.feature_names[sorted_indices]
sorted_importances = feature_importances[sorted_indices]

# Create a bar plot of the feature importances
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x=sorted_importances, y=sorted_feature_names, palette="viridis", legend=False) #need to specify the x and y in new seaborn
```
